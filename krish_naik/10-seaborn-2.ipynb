## Categorical plots
- BoxPlot
- Violinplot
- CountPlot
- Bar Plot

```python
import seaborn as sns

df = sns.load_dataset("tips")
```

```python
## Count plot

sns.countplot(data=df, x="sex")
```

```python
sns.countplot(df, y="sex")
```

```python
sns.countplot(data=df, x="smoker")
```

```python
sns.countplot(data=df, x="day")
```

```python
sns.barplot(df, y="total_bill", x="sex")
```

```python
## Box Plot
```

```python
sns.boxplot(df, x='sex', y="total_bill")
```
