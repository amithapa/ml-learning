```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple examples

x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
a = np.arange(40, 50)
b = np.arange(50, 60)
```

```python
## Plotting using matplotlib

## Scatter Plot
plt.scatter(x, y, c="r")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.savefig("tmp/test_scatter.png")
```

```python
y = x * x
```

```python
plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2d diagram")
```

```python
## Creating subtitles
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'g')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'b')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'r*', linestyle="dashed")
```

```python
x = np.arange(1, 11)
y = 3 * x + 5
plt.title("Matplotlib Demooo")
plt.xlabel("Label on X-Axis")
plt.ylabel("Label on Y-axis")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.title("sine wave form")

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)

plt.plot(x, y_sin, "r")
plt.title("Sin")

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, "g")
plt.title("Cosine")
```

```python
## Bar Plot
x = [2, 8, 10]
y = [11, 16, 9]

x2 = [3, 9, 11]
y2 = [6, 15, 7]

plt.bar(x, y)
plt.bar(x2, y2, color='g')
plt.title("Bar Graph example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

## Histograms

```python
a = np.array([29, 64, 47, 86, 57, 95, 46, 93, 79, 29, 41, 29, 68, 98, 85, 72, 35, 45, 47, 37, 68, 22, 61, 57, 77, 37, 53, 49, 30, 66])
plt.hist(a)
plt.title("histogram")
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=False)

plt.show()
```

## Pie Chart

```python
labels = "Python", "C++", "Rust", "Golang"
sizes = [215, 130, 245, 210]

colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)

plt.axis("Equal")
plt.show()
```
