```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

### Import the Boston Dataset

```python
house_price_dataset = pd.read_csv("dataset/housing.csv")
```

```python
house_price_dataset.head()
```

```python
house_price_dataset.shape
```

```python
# check for missing values
house_price_dataset.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataset.describe()
```

### Understanding the coorelation between various features in datasets

1. Positive coorelation
2. Negative coorelation

```python
coorelation = house_price_dataset.corr()
```

```python
# constructing heat map to understand the coorelation
plt.figure(figsize=(10, 10))

sns.heatmap(coorelation, cbar=True, square=True, fmt=".1f", annot=True, annot_kws={"size": 8}, cmap="Blues")
```

### Splitting the data and the target

```python
X = house_price_dataset.drop(["MEDV"], axis=1).values
y = house_price_dataset["MEDV"].values
```

```python
X.shape, y.shape
```

## Splitting the data into Training data and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

#### XGBoost Regressor

```python
# loading the model
classifier = XGBRegressor()
```

```python
# training the model with X_train
classifier.fit(X_train, y_train)
```

## Evaluation

#### Prediction on training data

```python
y_train_pred = classifier.predict(X_train)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, y_train_pred)

# mean absolute error
score_2 = metrics.mean_absolute_error(y_train, y_train_pred)

print(f"R squared error: {score_1}")

print(f"Mean absolute error: {score_2}")
```

### Visualizing the actual prices and predicted prices

```python
plt.scatter(y_train, y_train_pred)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual price vs Predicted price")
plt.show()
```

#### Prediction on testing data

```python
y_test_pred = classifier.predict(X_test)
metrics.r2_score(y_test, y_test_pred)
```

```python
# R squared error
score_1 = metrics.r2_score(y_test, y_test_pred)

# mean absolute error
score_2 = metrics.mean_absolute_error(y_test, y_test_pred)

print(f"R squared error: {score_1}")

print(f"Mean absolute error: {score_2}")
```

```python
plt.scatter(y_test, y_test_pred)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual price vs Predicted price")
plt.show()
```
