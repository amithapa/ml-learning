```python
breast_cancer_dataset.diagnosis.replace({"M": 1, "B": 0}, inplace=True)
```

## Feature & target extraction

```python
X = breast_cancer_dataset.drop(columns=["id", "diagnosis"], axis=1).values
y = breast_cancer_dataset["diagnosis"].values
```

## Splitting Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

## Training Model

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, y_train)
```

## Model evaluation

```python
# training data evaluation
y_train_predict = classifier.predict(X_train)

training_score = accuracy_score(y_train, y_train_predict)
training_score
```

```python
# training data evaluation
y_test_predict = classifier.predict(X_test)

testing_score = accuracy_score(y_test, y_test_predict)
testing_score
```
