## Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## Data loading

```python
# loading the data from the csv file to pandas dataframe
movie_dataset = pd.read_csv("dataset/movies.csv")
```

```python
movie_dataset.shape
```

```python
movie_dataset.head()
```

```python
# Selecting the relevant features for recommendation

selected_features = ["genres", "keywords", "title", "tagline", "cast", "director"]
print(selected_features)
```

```python
# movie_dataset[selected_features].fillna("")
for feature in selected_features:
    movie_dataset[feature] = movie_dataset[feature].fillna("")
```

```python
# combining all the 5 selected features

combined_features = movie_dataset["genres"] + " " + movie_dataset["keywords"] + " " + movie_dataset["title"] + " " + movie_dataset["tagline"] + " " + movie_dataset["cast"] + " " + movie_dataset["director"]
```

```python
combined_features
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

## Cosine Similarity

```python
simialrity = cosine_similarity(feature_vectors)
```

```python
simialrity
```

```python
simialrity.shape
```

```python
# getting the movie name from the user

movie_name = input("Enter your favorite movie name:")
```

```python
movie_name
```

```python
# creating a list with all the movie names given in the dataset
list_of_all_titles = movie_dataset["title"].tolist()
```

```python
print(list_of_all_titles)
```

```python
# finding the closematch for the movie name given by the user.

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = movie_dataset[movie_dataset.title == close_match]["index"].values[0]

print(index_of_the_movie)
```

```python
# getting a list of similar movies

simialrity_score = list(enumerate(simialrity[index_of_the_movie]))
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(simialrity_score, key=lambda k: k[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print("Movies suggested for you: ")

index = 1
for (movie_index, score) in sorted_similar_movies[:20]:
    title_from_index = movie_dataset[movie_dataset.index == movie_index].title.values[0]
    print(f"{index}. {title_from_index}")

    index += 1
```

## Movie recommendation system

```python
# getting the movie name from the user

movie_name = input("Enter your favorite movie name:")

# creating a list with all the movie names given in the dataset
list_of_all_titles = movie_dataset["title"].tolist()

# finding the closematch for the movie name given by the user.

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = find_close_match[0]

# finding the index of the movie with title
index_of_the_movie = movie_dataset[movie_dataset.title == close_match]["index"].values[0]

# getting a list of similar movies
simialrity_score = list(enumerate(simialrity[index_of_the_movie]))


# sorting the movies based on their similarity score
sorted_similar_movies = sorted(simialrity_score, key=lambda k: k[1], reverse=True)
print(sorted_similar_movies)

# print the name of similar movies based on the index

print("Movies suggested for you: ")

index = 1
for (movie_index, score) in sorted_similar_movies[:20]:
    title_from_index = movie_dataset[movie_dataset.index == movie_index].title.values[0]
    print(f"{index}. {title_from_index}")

    index += 1
```
